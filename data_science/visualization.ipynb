````python
x = np.arange(1,10*2,.2)
y = np.cos(x)
plt.plot(x,y)
plt.show()
````

````python
x = np.arange(1,10*2,.2)
y = np.cos(x)
plt.plot(x,y,'ro:',label='cos wave')
plt.legend()
plt.show()
````

````python
plt.figure(figsize=(13,4))
x = np.arange(1,10*2,.2)
y = np.cos(x)
plt.plot(x,y,'ro:',label='cos wave')
plt.legend()
plt.show()
````

````python
plt.figure(figsize=(13,4))
x = np.arange(1,10*2,.2)
y = np.cos(x)
y2 = np.sin(x)
plt.plot(x,y,'r:',label='cos wave')
plt.plot(x,y2,'g--',label='sine wave')
plt.legend()
plt.show()
````

````python
movies = [  
'The Lion King(1994)',
'Back to the Future(1985)',
'The Pianist(2002)',
'Terminator 2: Judgment Day(1991)',
'American History X(1998)',
'Modern Times(1936)',
'Psycho(1960)',
'Gladiator(2000)',
'City Lights(1931)',
'The Departed(2006)',
]
ratings = [
    9,
    5,
    6,
    9,
    4,
    8,
    6,
    8,
    3,
    6,
]
````

````python
plt.bar(x=movies,height=ratings)
plt.show()
````

````python
plt.figure(figsize=(7,3))
plt.bar(x=movies,height=ratings)
plt.xticks(rotation=90)
plt.show()
````

````python
plt.figure(figsize=(8,5))
plt.barh(y=movies,width=ratings)
plt.show()
````

````python
plt.figure(figsize=(7,3))
plt.bar(x=movies,height=ratings,hatch='//',color='red')
plt.xticks(rotation=90)
plt.savefig('moviebars.png',bbox_inches='tight')
````

````python
cities = ['lucknow','kanpur','jhansi','sitapur','barabanki','gonda']
rainfall = [200,100,60,211,288,150]
plt.pie(rainfall,labels=cities, autopct='%.1f')
plt.show()
````

````python
cities = ['lucknow','kanpur','jhansi','sitapur','barabanki','gonda']
rainfall = [200,100,60,211,288,150]
explode = [0,0,0.1,0,0,0]
plt.pie(rainfall,labels=cities, autopct='%.1f',explode=explode)
plt.show()
````

````python
cities = ['lucknow','kanpur','jhansi','sitapur','barabanki','gonda']
rainfall = [200,100,60,211,288,150]
explode = [0,0,0.1,0,0,0]
colors = ['#ea2c62','#ff9a8c','#fad5ad','#adb36e','#ddb36e','#adb3ff'] 
plt.pie(rainfall,
        labels=cities,
        autopct='%.1f',
        explode=explode, 
        shadow=True,
        colors=colors,
        )
plt.show()
````

````python
# donut chart
cities = ['lucknow','kanpur','jhansi']
rainfall = [200,100,60]
plt.pie(rainfall,labels=cities, autopct='%.1f', wedgeprops={'width':.3})
plt.show()
````

````python
# donut chart
cities = ['lucknow','kanpur','jhansi']
rainfall = [200,100,60]
plt.pie(rainfall,
        labels=cities,
        autopct='%.1f',
        wedgeprops={'width':.3}, 
        labeldistance=.4,
        pctdistance=1.1)
plt.savefig('rainfallpie.png',bbox_inches='tight')
````

## scatter plot
- show relation of 2 or 3 data values
- each values list must be  numeric and same in size

````python
x =  np.random.randint(1,100,100)
y = np.random.uniform(1,100,100)
````

````python
plt.plot(x,y)
plt.title("this is gibberish")
plt.show()
````

````python
x =  np.random.randint(1,100,100)
y = np.random.uniform(1,100,100)
plt.scatter(x,y,s=x*10,c=y,cmap='cool',marker='o',edgecolors='black',alpha=.4)
plt.title("this is information")
plt.show()
````

````python
import plotly.express as px
````

````python
px.scatter(x=x,y=y,color=y,size=x)
````

````python
x =  np.random.randint(1,100,100)
y = np.random.uniform(1,100,100)
z = np.random.uniform(1,100,100)
px.scatter_3d(x=x,y=y,z=z,color=y,size=x)
````
