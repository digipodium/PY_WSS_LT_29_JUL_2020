````python
x = np.arange(1,10*2,.2)
y = np.cos(x)
plt.plot(x,y)
plt.show()
````

````python
x = np.arange(1,10*2,.2)
y = np.cos(x)
plt.plot(x,y,'ro:',label='cos wave')
plt.legend()
plt.show()
````

````python
plt.figure(figsize=(13,4))
x = np.arange(1,10*2,.2)
y = np.cos(x)
plt.plot(x,y,'ro:',label='cos wave')
plt.legend()
plt.show()
````

````python
plt.figure(figsize=(13,4))
x = np.arange(1,10*2,.2)
y = np.cos(x)
y2 = np.sin(x)
plt.plot(x,y,'r:',label='cos wave')
plt.plot(x,y2,'g--',label='sine wave')
plt.legend()
plt.show()
````

````python
movies = [  
'The Lion King(1994)',
'Back to the Future(1985)',
'The Pianist(2002)',
'Terminator 2: Judgment Day(1991)',
'American History X(1998)',
'Modern Times(1936)',
'Psycho(1960)',
'Gladiator(2000)',
'City Lights(1931)',
'The Departed(2006)',
]
ratings = [
    9,
    5,
    6,
    9,
    4,
    8,
    6,
    8,
    3,
    6,
]
````

````python
plt.bar(x=movies,height=ratings)
plt.show()
````

````python
plt.figure(figsize=(7,3))
plt.bar(x=movies,height=ratings)
plt.xticks(rotation=90)
plt.show()
````

````python
plt.figure(figsize=(8,5))
plt.barh(y=movies,width=ratings)
plt.show()
````

````python
plt.figure(figsize=(7,3))
plt.bar(x=movies,height=ratings,hatch='//',color='red')
plt.xticks(rotation=90)
plt.show()
````
