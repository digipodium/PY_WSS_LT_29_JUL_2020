# Canada Immigration data analytics

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/Canada.xlsx', sheet_name='Canada by Citizenship', skiprows=range(20), skipfooter = 2)

df.head(5) # first 5 rows of data
```

```python
df.drop(columns=['Type','Coverage','AREA','REG','DEV'],inplace=True)
```

```python
df.head()
```

```python
df.rename(columns={'OdName':'Country','AreaName':'Continent','RegName':'Region'},inplace=True)
```

```python
df.head()
```

```python
df.set_index('Country',inplace=True)
```

```python
df
```

```python
df['total'] = df.sum(axis=1)
```

```python
# area plot
df.sort_values('total',ascending=False,inplace=True)
```

```python
df_top5 = df.head()
```

```python
df_top5 = df_top5[range(1980,2014)].T
```

```python
df_top5.head(2)
```

immigration country wise in top 5 countries

```python
plt.style.use('seaborn')
```

```python
df_top5.plot(kind='area',stacked=False, figsize=(20,10),alpha=.25)
plt.title('immigration country wise in top 5 countries')
plt.xlabel("Years")
plt.xlabel("Immgrants")
plt.show()
```
