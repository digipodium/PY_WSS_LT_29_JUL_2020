```python
import numpy as np
```

```python
nymap = folium.Map(location=newyork,zoom_start=10,tiles='stamen terrain')
complaints = folium.map.FeatureGroup()
for lat,lng in zip(df_100.Latitude,df_100.Longitude):
    if np.isnan(lat) or np.isnan(lng):
        continue
    complaints.add_child(
        folium.CircleMarker(
            [lat,lng],
            radius=5,
            color='red',
            fill=True,
            fill_color='blue',
            opacity=.6
        )
    )
nymap.add_child(complaints)
```

```python
nymap = folium.Map(location=newyork,zoom_start=10)
for lat,lng,label in zip(df_100.Latitude,df_100.Longitude,df_100['Complaint Type']):
    if np.isnan(lat) or np.isnan(lng):
        continue
    folium.Marker([lat,lng],popup=label).add_to(nymap)
nymap
```

```python
# for clusters in maps
from folium import plugins
```

```python
nymap = folium.Map(location=newyork,zoom_start=10)
cluster = plugins.MarkerCluster().add_to(nymap)
for lat,lng,label in zip(df_100.Latitude,df_100.Longitude,df_100['Complaint Type']):
    if np.isnan(lat) or np.isnan(lng):
        continue
    folium.Marker([lat,lng],popup=label).add_to(cluster)
nymap
```
